```python
# Distribution Analysis
# Close Value taqsimotini vizual tahlil qilish
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.histplot(sales_pipeline["close_value"],kde=True)
plt.title("Distribution of Close Value")
plt.show()
```

```python
# Revenue taqsimotini vizual tahlil qilish
plt.figure(figsize=(10,5))
sns.histplot(accounts["revenue"],kde=True)
plt.title("Distribution of Revenue")
plt.show()
```

```python
# Employees taqsimotini vizual tahlil qilish
plt.figure(figsize=(10,5))
sns.histplot(accounts["employees"],kde=True)
plt.title("Distribution of Employees")
plt.show()
```

```python
# Close Value ustunidagi outlierlarni aniqlash
plt.figure(figsize=(8,5))
sns.boxplot(x=sales_pipeline["close_value"])
plt.title("Outlier - Close Value")
plt.show()
```

```python
# Revenue ustunidagi outlierlarni aniqlash
plt.figure(figsize=(8,5))
sns.boxplot(x=accounts["revenue"])
plt.title("Outlier  - Revenue")
plt.show()
```

```python
# Umumiy close value qiymatini hisoblash
print("Total value: ",sales_pipeline["close_value"].sum())
```

```python
# Deal bosqichlari bo'yicha tahlil
print("Deal stage analysis: ",sales_pipeline["deal_stage"].value_counts())
```

```python
# Umumiy win rate ni hisoblash
won = (sales_pipeline["deal_stage"]=="Won").sum()
total = len(sales_pipeline)
win_rate = won/total*100
print("Win rate: ",win_rate)
```

```python
# Mahsulotlar bo'yicha umumiy close value ni hisoblash
sales_pipeline.groupby("product")["close_value"].sum()
```

```python
# Eng ko'p va eng kam daromad keltirgan mahsulotlarni aniqlash
product_revenue = sales_pipeline.groupby("product")["close_value"].sum()
print("Most revenue: ",product_revenue.sort_values(ascending=False).head(1))
print("")
print("Least revenue: ",product_revenue.idxmin())
```

```python
# Sales agentlar bo'yicha umumiy close value ni hisoblash
sales_pipeline.groupby("sales_agent")["close_value"].sum()
```

```python
# Eng yaxshi va eng past natijali sales agentlarni aniqlash
print("Top agent: ",sales_pipeline.groupby("sales_agent")["close_value"].sum().idxmax())
print("Worst agent: ",sales_pipeline.groupby("sales_agent")["close_value"].sum().idxmin())
```

```python
# Oylik daromad trendini tahlil qilish
sales_pipeline["close_date"] = pd.to_datetime(sales_pipeline["close_date"])
monthly_sales = (sales_pipeline.groupby(sales_pipeline["close_date"].dt.to_period("M"))["close_value"].sum())
monthly_sales.plot(figsize=(12,5))
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```
